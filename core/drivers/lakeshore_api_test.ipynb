```python
# -*- coding: utf-8 -*-
"""
Example: read channel old measurement data.
"""
# --- Imports
#
import json
import requests
# --- Setup
# Define your device IP here
DEVICE_IP = '127.0.0.1'
# Timeout for get/post operations (in seconds)
TIMEOUT = 100
# --- Read channel N old measurement data in defined time period
#
url = 'http://{}:49099/values'.format(DEVICE_IP)
data = {
 'channel_nr': 8,
 'start_time': '2019-10-01T00:00:00Z',
 'stop_time': '2019-10-02T00:00:00Z',
 'fields': [
 'temperature',
 'resistance',
 'reactance'
 ]
}
#
req = requests.post(url, json=data, timeout=TIMEOUT)
data = req.json()
#
print('Response: \n{}'.format(json.dumps(data, indent=2)))
```
