```python
import os, json
from json.decoder import JSONDecodeError
working_directory = os.getcwd()

my_dict = {'testf': {'nesty shit': 2334},
           'testgf':1}

from os import getcwd
from json.decoder import JSONDecodeError
from json import load, dump

def dump_dict(dict):
    dir = f"{getcwd()}//config.json"
    try:
        with open(dir, 'r') as f:
            try:
                my_dict = load(f)
            except JSONDecodeError:
                my_dict = {}
    except FileNotFoundError:
        my_dict = {}

    for k in dict.keys():
        my_dict[k] = dict[k]

    with open(f"{working_directory}//config.json", 'w') as f:
        dump(my_dict, f, indent=4)

dump_dict(my_dict)

 
```
